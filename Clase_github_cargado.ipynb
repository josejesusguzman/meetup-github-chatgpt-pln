```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

def load_model(model_name='gpt2'):
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)
    model = GPT2LMHeadModel.from_pretrained(model_name)
    return tokenizer, model

def generate_text(input_text, tokenizer, model):
    inputs = tokenizer.encode(input_text, return_tensors='pt')
    outputs = model.generate(inputs, max_length=150, num_return_sequences=1, no_repeat_ngram_size=2, pad_token_id=tokenizer.eos_token_id)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

tokenizer, model = load_model()
print(generate_text("Hello, how are you?", tokenizer, model))
```

# Vectorización

Imagina que estás en una biblioteca con muchos libros y quieres encontrar un libro que hable sobre gatos. Podrías empezar a buscar todos los libros que usen la palabra "gato", pero te encontrarías con un problema: algunos libros, como un diccionario, mencionan la palabra "gato" aunque no hablen realmente sobre gatos.

Aquí es donde el TfidfVectorizer nos puede ayudar. Es como un detective súper inteligente que no solo busca los libros que usan la palabra "gato", sino que también se da cuenta de qué tan importante es esa palabra para cada libro.

El TfidfVectorizer utiliza dos ideas principales:

TF (Term Frequency): Cuántas veces aparece una palabra en un libro. Por ejemplo, si un libro menciona "gato" muchas veces, es más probable que ese libro hable sobre gatos.

IDF (Inverse Document Frequency): Qué tan común es una palabra en todos los libros de la biblioteca. Si la palabra "gato" aparece en casi todos los libros, entonces no es una buena palabra para distinguir los libros que realmente hablan sobre gatos de los que no lo hacen.

Entonces, el TfidfVectorizer toma estas dos ideas y las combina para darle a cada palabra en cada libro una puntuación. Las palabras que aparecen a menudo en un libro, pero raramente en todos los demás libros, obtienen una puntuación alta. Estas son las palabras que probablemente son muy importantes para ese libro.

Al final, gracias al TfidfVectorizer, puedes encontrar los libros que realmente hablan sobre gatos, y no solo los que mencionan la palabra "gato" un par de veces.

```python
# Vectorización
documento_a = 'el perro es de color negro'
documento_b = 'ponchi es el nombre de mi perro'
documento_c = 'no puedo tener mascotas en mi casa'
documento_d = 'tu perro es más grande que mi perro'
```

# Lematización

La lematización en el Procesamiento del Lenguaje Natural (PLN) es un proceso que transforma una palabra en su forma base o raíz, llamada "lema".

Por ejemplo, si tienes las palabras "correr", "corrió", "corriendo", todas estas palabras están en diferentes formas, pero su lema es "correr". Así que la lematización las convertiría todas a "correr".

Esto es útil en muchos casos de PLN, como la búsqueda de texto y el análisis de sentimientos, porque nos permite tratar palabras diferentes que tienen el mismo significado básico como si fueran la misma palabra.

Es importante notar que la lematización es un proceso más complejo que simplemente eliminar las terminaciones de las palabras (lo que se llama "stemming"). La lematización realmente entiende el contexto en el que se usa una palabra para convertirla en su forma base. Por ejemplo, el lema de "mejor" podría ser "bueno" dependiendo del contexto, algo que un proceso de stemming no podría capturar.

```python
def clean(text):
    wn = nltk.WordNetLemmatizer()
    stopword = nltk.corpus.stopwords.words('english')
    tokens = nltk.word_tokenize(text)
    lower = [word.lower() for word in tokens]
    no_stopwords = [word for word in lower if word not in stopword]
    no_alpha = [word for word in no_stopwords if word.isalpha()]
    lemm_text = [wn.lemmatize(word) for word in no_alpha]
    clean_text = lemm_text
    return clean_text
```
